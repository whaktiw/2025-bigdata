```python
import requests
url ='http://openapi.seoul.go.kr:8088/44785558686d69723837666d796941/json/viewNightSpot/1/100/'
response = requests.get(url)
print(response.content)
```

```python
data =response.json()
data
```

```python
import pandas as pd
temp=[]
for d in data['viewNightSpot']['row']:
  temp.append(d)

uni = pd.DataFrame(temp)
uni.head()
```

```python
import folium
anMap = folium.Map(location=[37.51384422, 127.02710693],
                   zoom_start=11,
                   zoom_control=False,
                   control_scale=True)
anMap
```

```python
night2 = folium.FeatureGroup(name='night2')
anMap.add_child(night2)
```

```python

from folium.plugins import MarkerCluster
mc = MarkerCluster().add_to(night2)

for n, row in uni.iterrows():
  folium.Marker(location=[row['LA'], row['LO']], popup = folium.Popup(row['ADDR'], max_width=200)).add_to(mc)

anMap
```
